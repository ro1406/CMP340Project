```python

def binarySearch(arr, l, r, x):
    count = 0
    while l <= r:
                
        mid = (l+r) // 2;
        if arr[mid] == x: # check if value of x is in the middle
            return count+1
        elif arr[mid] < x:  # search right
            l = mid + 1
        else:
            r = mid - 1 # search left
        count=count+1
    return count # marks that element was not found.
```

```python
def interpolation_search(ordered_list, term):
    count = 0
    size_of_list = len(ordered_list) - 1

    index_of_first_element = 0
    index_of_last_element = size_of_list

    while index_of_first_element <= index_of_last_element:
        
        mid_point = index_of_first_element + (index_of_last_element - index_of_first_element) * (term - ordered_list[index_of_first_element]) // (ordered_list[index_of_last_element] - ordered_list[index_of_first_element])


        if mid_point > index_of_last_element or mid_point < index_of_first_element:
            return count+1

        if ordered_list[mid_point] == term:
            return count+1

        if term > ordered_list[mid_point]:
            index_of_first_element = mid_point + 1
        else:
            index_of_last_element = mid_point - 1
        count = count+1
    if index_of_first_element > index_of_last_element:
        return count+1
```
